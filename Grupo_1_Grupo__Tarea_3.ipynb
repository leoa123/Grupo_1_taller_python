Utilizando scipy. stats, crea un programa que modele la probabilidad de obtener un
número específico de éxitos en una serie de lanzamientos de una moneda sin trucos
(probabilidad de éxito p = 0. 5). Genera la probabilidad de obtener exactamente 7 caras
en 10 lanzamientos

```python
from scipy.stats import binom

# Parámetros
n = 10  # número de lanzamientos
p = 0.5  # probabilidad de éxito en un solo lanzamiento
k = 7  # número de caras/sellos

# Cálculo de la probabilidad
probabilidad = binom.pmf(k, n, p)
print(f"La probabilidad de obtener {k} caras en {n} lanzamientos es: {probabilidad}")
```

II. Calcula la probabilidad acumulada de obtener 7 o menos caras en 10 lanzamientos

```python
probabilidad_acumulada = binom.cdf(k, n, p)
print(f"La probabilidad acumulada de obtener {k} o menos caras en {n} lanzamientos es: {probabilidad_acumulada}")
```

V. Utilizando scipy. stats, realiza un análisis de una distribución normal con una media de
0 y desviación estándar de 1. Genera un conjunto de datos aleatorios con 1000
muestras a partir de esta distribución.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
media = 0
desviacion_estandar = 1
n_muestras = 1000

datos_random = norm.rvs(loc=media, scale=desviacion_estandar, size=n_muestras)
plt.hist(datos_random, bins=30, density=True, alpha=0.6, color='g')
plt.show()
```

VI. Calcula y muestra la media, la desviación estándar y la mediana de los datos
generados.

```python
media_obtenida = np.mean(datos_random)
desviacion_estandar_obtenida = np.std(datos_random)
mediana_obtenida = np.median(datos_random)
print(f"Media obtenida: {media_obtenida}")
print(f"Desviación estándar obtenida: {desviacion_estandar_obtenida}")
print(f"Mediana obtenida: {mediana_obtenida}")
```

VII. Grafica el histograma de los datos y superpón la curva de la distribución normal
teórica.

```python
datos_random = norm.rvs(loc=media, scale=desviacion_estandar, size=n_muestras)
plt.hist(datos_random, bins=30, density=True, alpha=0.6, color='g')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, media, desviacion_estandar)
plt.plot(x, p, 'k', linewidth=2)

# Título y etiquetas del gráfico
plt.title('Histograma de datos generados y curva de distribución normal')
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)

plt.show()
```

VIII.Calcula la probabilidad de que un valor caiga dentro del rango de -1 a 1 (una desviación
estándar de la media).

```python
probabilidad = norm.cdf(1, loc=media, scale=desviacion_estandar) - norm.cdf(-1, loc=media, scale=desviacion_estandar)

print(f"La probabilidad de que un valor caiga dentro del rango de -1 a 1 es: {probabilidad}")
```
